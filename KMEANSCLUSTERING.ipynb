```python
import pandas as pd
import pickle
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv(r"C:\Users\viswa\OneDrive\Desktop\task_8\Simple RFM Customer Segmentation - K Means Practice.csv")

print("Columns:", df.columns)

# Select only numeric columns (SAFE METHOD)
X = df.select_dtypes(include=["int64", "float64"])

print("Using features:", X.columns)

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train KMeans
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(X_scaled)

# Assign clusters
df["Cluster"] = kmeans.labels_

# Save model and scaler
with open("kmeans_model.pkl", "wb") as f:
    pickle.dump(kmeans, f)

with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

print("✅ KMeans model & scaler created successfully")
df.head()
```
